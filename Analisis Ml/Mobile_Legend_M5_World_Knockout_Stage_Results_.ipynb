```python

# Function to convert 'K' values to numeric
def convert_to_numeric(value):
    if isinstance(value, str) and 'K' in value:
        return float(value.replace('K', '')) * 1000
    else:
        return float(value)

# Apply the function to the 'Gold' column
Game['Gold'] = Game['Gold'].apply(convert_to_numeric)

# Now, the 'Gold' column contains numeric values without 'K'
print(Game)
```

```python
import pandas as pd
import plotly.graph_objects as go

# Assuming you have your DataFrame 'Game' with the specified columns

teams_of_interest = [
    'AP.Bren',
    'Blacklist Intl.',
    'ONIC Esports',
    'Geek Fam ID',
    'See You Soon',
    'Fire Flux Esports',
    'Burmese Ghouls'
    # Add other teams as needed
]

# Filter data for the specified teams
filtered_data = Game[Game['Team'].isin(teams_of_interest)]

# Create a function to get the top banned or picked heroes for a given team
def get_top_heroes(team, action):
    return (
        filtered_data[filtered_data['Team'] == team]
        .groupby([action, 'Role ' + action])
        .size()
        .reset_index(name='Count')
        .sort_values(by='Count', ascending=False)
        .head(5)
    )

# Create an interactive scatter plot using Plotly for banned and picked heroes
def create_hero_plot(fig, action, title):
    for team in teams_of_interest:
        top_heroes_team = get_top_heroes(team, action)
        fig.add_trace(
            go.Scatter(
                x=top_heroes_team[action],
                y=top_heroes_team['Count'],
                mode='markers',
                name=f'{team} - Top {action.capitalize()}ed Heroes',
                text=top_heroes_team['Role ' + action] + ' - ' + top_heroes_team['Count'].astype(str),
                marker=dict(size=12)
            )
        )

    fig.update_layout(
        title=f'Top {action.capitalize()}ed Heroes for Each Team',
        xaxis_title='Hero',
        yaxis_title='Count',
        showlegend=True
    )

# Create figures for both banned and picked heroes
fig_ban = go.Figure()
create_hero_plot(fig_ban, 'Ban', 'Banned')

fig_pick = go.Figure()
create_hero_plot(fig_pick, 'Pick', 'Picked')

# Show the interactive plots
fig_ban.show()
fig_pick.show()
```

# **Distribution Gold, Kill, Death and Asist each Team in each match game**

```python
import pandas as pd
import plotly.graph_objects as go
import dash
from dash import dcc, html
from dash.dependencies import Input, Output

# Assuming you have your DataFrame 'Game' with the specified columns
# Replace 'YOUR_CSV_FILE.csv' with the actual file or DataFrame

# Create a Dash web application
app = dash.Dash(__name__)

# Define the layout of the web app
app.layout = html.Div([
    html.Label("Select Team:"),
    dcc.Dropdown(
        id='team-dropdown',
        options=[{'label': team, 'value': team} for team in Game['Team'].unique()],
        value=Game['Team'].unique()[0],  # Default value
        multi=False
    ),

    dcc.Graph(id='gold-plot'),
    dcc.Graph(id='kills-plot'),
    dcc.Graph(id='deaths-plot'),
    dcc.Graph(id='assists-plot')
])

def create_team_plot(data, team, stat, title, yaxis_title):
    team_data = data[data['Team'] == team]
    fig = go.Figure()
    for player in team_data['Players'].unique():
        player_data = team_data[team_data['Players'] == player]
        fig.add_trace(
            go.Scatter(
                x=player_data.apply(lambda row: f"{row['Match']} - {row['Date']}", axis=1),
                y=player_data[stat],
                mode='lines+markers',
                name=f'{player} - {stat}',
                text=player_data['Pick'],  # Assuming 'Pick' contains hero names
            )
        )
    fig.update_layout(
        title=title,
        xaxis_title='Date - Match',
        yaxis_title=yaxis_title,
        showlegend=True
    )
    return fig

# Callbacks to update the figures based on team selection
@app.callback(
    [Output('gold-plot', 'figure'),
     Output('kills-plot', 'figure'),
     Output('deaths-plot', 'figure'),
     Output('assists-plot', 'figure')],
    [Input('team-dropdown', 'value')]
)
def update_figures(selected_team):
    gold_fig = create_team_plot(Game, selected_team, 'Gold', f'Gold Over Matches - {selected_team}', 'Gold')
    kills_fig = create_team_plot(Game, selected_team, 'K', f'Kills Over Matches - {selected_team}', 'Kills')
    deaths_fig = create_team_plot(Game, selected_team, 'D', f'Deaths Over Matches - {selected_team}', 'Deaths')
    assists_fig = create_team_plot(Game, selected_team, 'A', f'Assists Over Matches - {selected_team}', 'Assists')
    return gold_fig, kills_fig, deaths_fig, assists_fig

# Run the web app
if __name__ == '__main__':
    app.run_server(debug=True)
```
